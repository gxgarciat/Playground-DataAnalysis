# Multiple minima vs Initial Guess

## $$g(x) = x^4 -4x^2 +5$$

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Data generation
x_1 = np.linspace(start=-2,stop=2,num=1000)

# Definition of functions
def func_g(x):
    return (x**4 - 4*(x**2) + 5)

def dg(x):
    return 4*(x**3) - 8*x 
```

```python
# Plotting equation and derivative, next to each other

# Scaling the graph
plt.figure(figsize=[15,5])

# 1 Chart: Cost function
plt.subplot(1,2,1) # This command decides how to stack the two graphs. If required, (2,1,1), to have them on top of each other

plt.xlim([-3,3])
plt.ylim(0,8)
plt.grid(True)
plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.title('Cost Function')
plt.xlabel('X',fontsize=16)
plt.ylabel('g(x)',fontsize=16)
plt.plot(x_1,func_g(x_1),'r')

# 1 Chart: Derivative function
plt.subplot(1,2,2)
plt.title('Slope of the function')
plt.xlabel('X',fontsize=16)
plt.ylabel('dg(x)',fontsize=16)
plt.grid(True)

plt.plot(x_1,dg(x_1),'b')
plt.show()
```

```python
def gradient_descent(derivative_fun, initial_guess, learning_rate, precision):
    
    new_x = initial_guess
    x_list = [new_x]
    #slope_list = [derivative_func(new_x)]
    slope_list = [derivative_fun(new_x)]

    # for loop
    # steps = 50
    #steps = 500

    for n in range(500):
        previous_x = new_x
        gradient = derivative_fun(previous_x)
        new_x = previous_x - learning_rate * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_fun(new_x))
        
        if step_size < precision:
            break
        
    return new_x, x_list, slope_list
```

```python
# Finding +X
local_min, list_x, deriv_list = gradient_descent(derivative_fun = dg, initial_guess = 0.5, learning_rate = 0.02, precision = 0.001)
print('Local min occurs at:', local_min)
print('Number of steps:', len(list_x))
```

```python
# Finding -X
local_min, list_x, deriv_list = gradient_descent(derivative_fun = dg, initial_guess = -0.5, learning_rate = 0.02, precision = 0.001)
print('Local min occurs at:', local_min)
print('Number of steps:', len(list_x))
```

```python
# Plotting function, scatter and derivative, next to each other

local_min, list_x, deriv_list = gradient_descent(derivative_fun = dg, initial_guess = 0.1, learning_rate = 0.02, precision = 0.001)


# Scaling the graph
plt.figure(figsize=[15,5])

# 1 Chart: Cost function
plt.subplot(1,2,1) # This command decides how to stack the two graphs. If required, (2,1,1), to have them on top of each other

plt.xlim([-3,3])
plt.ylim(0,8)
plt.grid(True)
plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.title('Cost Function')
plt.xlabel('X',fontsize=16)
plt.ylabel('g(x)',fontsize=16)
plt.plot(x_1,func_g(x_1),'r')
plt.scatter(list_x,func_g(np.array(list_x)),s=100,color='green',alpha=0.6)

# 1 Chart: Derivative function
plt.subplot(1,2,2)
plt.title('Slope of the function')
plt.xlabel('X',fontsize=16)
plt.ylabel('dg(x)',fontsize=16)
plt.grid(True)

plt.plot(x_1,dg(x_1),'b')
plt.scatter(list_x,deriv_list,s=100,color='green',alpha=0.6)

plt.show()
```
