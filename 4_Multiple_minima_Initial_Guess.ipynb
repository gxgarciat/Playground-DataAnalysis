# Multiple minima vs Initial Guess

## $$g(x) = x^4 -4x^2 +5$$

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Data generation
x_1 = np.linspace(-2,2,1000)

# Definition of functions
def func_g(x):
    return (x**4 - 4*(x**2) + 5)

def dg(x):
    return 4*(x**3) - 8*x 
```

```python
# Plotting equation and derivative, next to each other

# Scaling the graph
plt.figure(figsize=[15,5])

# 1 Chart: Cost function
plt.subplot(1,2,1) # This command decides how to stack the two graphs. If required, (2,1,1), to have them on top of each other

plt.xlim([-3,3])
plt.ylim(0,8)
plt.grid(True)
plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.title('Cost Function')
plt.xlabel('X',fontsize=16)
plt.ylabel('g(x)',fontsize=16)
plt.plot(x_1,func_g(x_1),'r')

# 1 Chart: Derivative function
plt.subplot(1,2,2)
plt.title('Slope of the function')
plt.xlabel('X',fontsize=16)
plt.ylabel('dg(x)',fontsize=16)
plt.grid(True)

plt.plot(x_1,dg(x_1),'b')
plt.show()
```

```python
def gradient_descent(derivative_fun, initial_guess, multiplier, precision):
    
    new_x = initial_guess
    x_list = [new_x]
    #slope_list = [derivative_func(new_x)]
    slope_list = [dg(new_x)]

    # for loop
    # steps = 50
    steps = 100

    for n in range(steps):
        previous_x = new_x
        gradient = dg(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)

        x_list.append(new_x)
        #slope_list.append(dg(new_x))
        slope_list.append(dg(new_x))
        
        if step_size < precision:
            break
        
        return new_x, x_list,slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(dg, 0.5, 0.02, 0.001)
print(f'Local minimum is: {local_min}')    
print('Number of steps: ', len(list_x))
```
