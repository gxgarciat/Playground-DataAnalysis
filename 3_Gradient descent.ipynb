# Notebook imports and packages

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Cost function
$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Data generation based on the equation
x_1 = np.linspace(start=-3,stop=3,num=100)
#x_1
```

```python
# Plotting equation
plt.xlim([-3,3])
plt.ylim(0,8)
plt.grid(True)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1,f(x_1))
plt.show()
```

# Derivates

```python
def df(x):
    return 2*x +1
```

```python
# Plotting equation and derivative, next to each other

# Scaling the graph
plt.figure(figsize=[15,5])

# 1 Chart: Cost function
plt.subplot(1,2,1) # This command decides how to stack the two graphs. If required, (2,1,1), to have them on top of each other

plt.xlim([-3,3])
plt.ylim(0,8)
plt.grid(True)
plt.title('Cost Function')
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1,f(x_1),'r')

# 1 Chart: Derivative function
plt.subplot(1,2,2)
plt.title('Slope of the function')
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.grid(True)

plt.plot(x_1,df(x_1),'b')
plt.show()
```

# Gradient descent of the function $f(x) = x^2 + x + 1$

```python
new_x = 3
previous_x = 0
step_multiplier = 0.1

x_list = [new_x]
slope_list = [df(new_x)]

# for loop
# steps = 50
steps = 100

for n in range(steps):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    step_size = abs(new_x - previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))

print(f'Local minimum is: {new_x} and the slope at this point is {df(new_x)}')    
print('f(x) value at this point: ', f(new_x))
```

```python
# Plotting derivative approx. - gradient descent calculation

# Scaling the graph
plt.figure(figsize=[15,5])

# 1 Chart: Cost function
plt.subplot(1,2,1) # This command decides how to stack the two graphs. If required, (2,1,1), to have them on top of each other

plt.xlim([-3,3])
plt.ylim(0,8)
plt.grid(True)
plt.title('Cost Function')
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1,f(x_1),'g')

values = np.array(x_list)  # converting the list into an array
plt.scatter(x_list,f(values),color='red')

# 1 Chart: Derivative function
plt.subplot(1,2,2)
plt.title('Slope of the function')
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.grid(True)

plt.scatter(x_list, slope_list, color = 'red')
plt.plot(x_1,df(x_1),'b')
plt.show()
```
