## Correlation

## $$ \rho _{XY} = corr(X,Y)$$
## $$ -1.0 \leq \rho _{XY} \leq + 1.0 $$

```python
# Correlation between price and numer of rooms

data['PRICE'].corr(data['RM'])
```

```python
# Correlation between price and pupil teacher ratio

data['PRICE'].corr(data['PTRATIO'])
```

```python
# Total correlation of datasets - Pearson Correlation
# Pearson is only valid for continuous data. It is not valid for dummy variables

data.corr()
```

```python
# Processing correlation with Numpy

mask = np.zeros_like(data.corr())
triangle_indices = np.triu_indices_from(mask)
mask[triangle_indices] = True
#mask
```

```python
# Creating a heatmap

plt.figure(figsize=(16,10))
sns.heatmap(data.corr(),mask=mask,annot=True, annot_kws={"size":14})
sns.set_style('white')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
# Analyzing data - pollution and distance

plt.scatter(x=data['DIS'],y=data['NOX'])

plt.title('DIST vs NOX')
plt.xlabel('Distance from employment centre',fontsize=14)
plt.ylabel('NOX',fontsize=14)
plt.show()
```
