## p-values 

```python
X_incl_const = sm.add_constant(X_train)

model = sm.OLS(y_train,X_incl_const)

results = model.fit()

#results.params
#results.pvalues

pd.DataFrame({'coef': results.params,'p-value':round(results.pvalues,3)})

# p-values over 0.05 are not significant
```

## Testing for multicollinearity - VIF (Variance Inflation factor)

## $$ TAX = \alpha _0 + \alpha _1 RM + \alpha _2 NOX + ... \alpha _{12}LSTAT $$

## $$ VIF _{TAX} = \frac{1} {(1 - R _{TAX}^2)} $$

```python
variance_inflation_factor(exog = X_incl_const.values,exog_idx = 1) # .values will transform DataFrame into an n_array
```

```python
# Gathering number of columns for the for_loop

# len(X_incl_const.columns)
# a,b = X_incl_const.shape
# X_incl_const.shape[0] # rows
# X_incl_const.shape[1] # columns
```

```python
vif = []

for i in range(X_incl_const.shape[1]):
    vif.append(variance_inflation_factor(exog = X_incl_const.values,exog_idx = i))

# type(vif)
print(vif)
```

```python
# Second method

vif = [variance_inflation_factor(exog = X_incl_const.values,exog_idx = i) for i in range(X_incl_const.shape[1])]

# print(vif)
pd.DataFrame({'coef_name':X_incl_const.columns,'vif':np.round(vif,2)})
```

## Model Simplification & BIC

```python
# Original model - All features included

X_incl_const = sm.add_constant(X_train)

model = sm.OLS(y_train,X_incl_const)

results = model.fit()

org_coef = pd.DataFrame({'coef': results.params,'p-value':round(results.pvalues,3)})

# Obtaining BIC & r-squared

print('BIC is', results.bic)
print('r-squared is', results.rsquared)
```

```python
# Reduced model - INDUS excluded

X_incl_const = sm.add_constant(X_train)
X_incl_const = X_incl_const.drop(['INDUS'],axis=1)

model = sm.OLS(y_train,X_incl_const)

results = model.fit()

coef_minus_indus = pd.DataFrame({'coef': results.params,'p-value':round(results.pvalues,3)})

# Obtaining BIC & r-squared

print('BIC is', results.bic)
print('r-squared is', results.rsquared)
```

## Analysis
## After removing INDUS, abs(BIC) becomes bigger. 
## Our model has improved with an r-squared kept at the same level

```python
# Reduced model - INDUS & AGE excluded

X_incl_const = sm.add_constant(X_train)
X_incl_const = X_incl_const.drop(['INDUS'],axis=1)
X_incl_const = X_incl_const.drop(['AGE'],axis=1)

model = sm.OLS(y_train,X_incl_const)

results = model.fit()

reduc_coef = pd.DataFrame({'coef': results.params,'p-value':round(results.pvalues,3)})

# Obtaining BIC & r-squared

print('BIC is', results.bic)
print('r-squared is', results.rsquared)
```

## Analysis
## After removing INDUS and AGE, abs(BIC) becomes bigger. 
## Our model has improved with an r-squared kept at the same level

## Residuals & Residuals plots

```python
# Modified model: Transformed (using log prices) & simplified. This model has dropped two features

prices = np.log(data['PRICE']) # Using log prices
features = data.drop(['PRICE','INDUS','AGE'],axis=1)

# Tuple unpacking
X_train, X_test, y_train, y_test = train_test_split(features,prices,test_size=0.2,random_state=10) # Shuffling the data - 80/20 split

# Using Stasmodel
X_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train,X_incl_const)
results = model.fit()

# Residuals
residuals = y_train - results.fittedvalues
residuals

# type(residuals)

# Mean Squared Error

reduced_log_mse = round(results.mse_resid,3)
reduced_log_rsquared = round(results.rsquared,3)
```

```python
residuals.describe()
```

```python
# Alternative method

results.resid
```

```python
# Comparison between Predicted and Actual values

corr = round(y_train.corr(results.fittedvalues),2)
corr

plt.scatter(x=y_train,y=results.fittedvalues,c='indigo',alpha=0.6)
plt.plot(y_train,y_train)

plt.xlabel('Actual log prices $y _i$',fontsize=14)
plt.ylabel('Predicted log prices $\hat y _i$',fontsize=14)
plt.title(f'Actual vs Predicted log prices: $y _i$ vs $\hat y_i$ (Corr {corr})',fontsize=17)

plt.show()


plt.scatter(x=np.e**y_train,y=np.e**results.fittedvalues,c='red',alpha=0.6)
plt.plot(np.e**y_train,np.e**y_train)

plt.xlabel('Actual prices 000s $y _i$',fontsize=14)
plt.ylabel('Predicted prices 000s $\hat y _i$',fontsize=14)
plt.title(f'Actual vs Predicted prices: $y _i$ vs $\hat y_i$ (Corr {corr})',fontsize=17)

plt.show()

# Residuals vs predicted values


plt.scatter(x=results.fittedvalues,y=results.resid,c='blue',alpha=0.6)

plt.xlabel('Predicted log prices $\hat y _i$',fontsize=14)
plt.ylabel('Residuals',fontsize=14)
plt.title(f'Residuals vs fitted values $y _i$ vs $\hat y_i$ (Corr {corr})',fontsize=17)

plt.show()

```

```python
# Distribution of residuals (log prices) - Assessing normality

resid_mean = round(results.resid.mean(),3)
resid_skew = round(results.resid.skew(),3)

sns.displot(results.resid,color='navy')
plt.title(f'Log price model: residuals skew ({resid_skew}) - Mean ({resid_mean})')
plt.show()
```

```python
# Model ommiting features

prices = np.log(data['PRICE']) 
features = data.drop(['PRICE','INDUS','AGE','LSTAT','RM','NOX','CRIM'],axis=1)

# Tuple unpacking
X_train, X_test, y_train, y_test = train_test_split(features,prices,test_size=0.2,random_state=10) # Shuffling the data - 80/20 split

# Using Stasmodel
X_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train,X_incl_const)
results = model.fit()

# Comparison between Predicted and Actual values

corr = round(y_train.corr(results.fittedvalues),2)

plt.scatter(x=y_train,y=results.fittedvalues,c='black',alpha=0.6)
plt.plot(y_train,y_train)

plt.xlabel('Actual log prices $y _i$',fontsize=14)
plt.ylabel('Predicted log prices  $\hat y _i$',fontsize=14)
plt.title(f'Actual vs Predicted prices with ommited variables: $y _i$ vs $\hat y_i$ (Corr {corr})',fontsize=17)

plt.show()


# Residuals vs predicted values


plt.scatter(x=results.fittedvalues,y=results.resid,c='black',alpha=0.6)

plt.xlabel('Predicted prices $\hat y _i$',fontsize=14)
plt.ylabel('Residuals',fontsize=14)
plt.title(f'Residuals vs fitted',fontsize=17)

plt.show()


# Mean Squared Error

omitted_var_mse = round(results.mse_resid,3)
omitted_var_rsquared = round(results.rsquared,3)
```
